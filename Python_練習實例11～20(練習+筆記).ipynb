Python 练习实例13
```
题目：打印出所有的"水仙花数"，所谓"水仙花数"是指一个三位数，其各位数字立方和等于该数本身。
例如：153是一个"水仙花数"，因为153=1的三次方＋5的三次方＋3的三次方。
```

````
# myself test

import math

for i in range(100, 1000):
  hundred = i / 100    # 浮點數
  ten = (i % 100) / 10  # 浮點數
  unit = i % 10
  print(hundred, ten, unit)
  if (math.pow(hundred, 3) + math.pow(ten, 3) + math.pow(unit, 3)) == i:
   print(i)
````

````
# myself 

import math

for i in range(100, 1000):
  hundred = i // 100    # 浮點數
  ten = (i % 100) // 10  # 浮點數
  unit = i % 10
  if (math.pow(hundred, 3) + math.pow(ten, 3) + math.pow(unit, 3)) == i:
   print(i)
````

````
# answer

for n in range(100,1000):
    i = n // 100
    j = n // 10 % 10
    k = n % 10
    if n == i*i*i + j*j*j + k*k*k: 
        print(n)
````

Python 练习实例14
```
题目：将一个正整数分解质因数。例如：输入90,打印出90=2*3*3*5。
```

````
# myself

num = eval(input())
prime = [2, 3, 5, 7, 11, 13, 17, 19]
ans = []
for i in prime:     # 第一次後的迴圈沒有從list第一個開始
  if num % i == 0:
    num //= i
    ans.append(i)
ans
````
